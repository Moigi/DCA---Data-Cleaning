```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel('/content/drive/MyDrive/Data/staff (1).xlsx')
df.head()
```

## Converting categorical to numeric

```python
df['Left'].replace(['Active', 'Left'],
                        [0, 1], inplace=True)
data = df
```

```python
df.head()
```

```python
# count plot on single categorical variable
ax = sns.countplot(y ='Left', data = data, palette='colorblind')
# get the total count of the type column
total = df['Left'].count()

# annotate the bars with fmt from matplotlib v3.7.0
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')

# add space at the end of the bar for the labels
ax.margins(x=0.1)

ax.set(xlabel='Count', ylabel='Left', title='Employee Turnover Rate')
plt.show()
```

```python
def percentage_above_bar_relative_to_xgroup(ax):
    all_heights = [[p.get_height() for p in bars] for bars in ax.containers]
    for bars in ax.containers:
        for i, p in enumerate(bars):
            total = sum(xgroup[i] for xgroup in all_heights)
            percentage = f'{(100 * p.get_height() / total) :.1f}%'
            ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()), size=11, ha='center', va='bottom')


plt.figure(figsize=(12, 8))
ax3 = sns.countplot(x="department", hue="Left", data=data)
ax3.set(xlabel='Department', ylabel='Count')
plt.title("Employee Attrition (Left) by Department")
percentage_above_bar_relative_to_xgroup(ax3)
plt.show()
```
