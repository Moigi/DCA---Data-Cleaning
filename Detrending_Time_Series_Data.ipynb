```r
# All the necessary packages
library(tidyverse)
library(zoo)
library(xts)
library(readxl)
library(readr)
```

# i. Condensed Data dataset
**Cap Minimum before detrending**

```r
# All the datasets
Condensed_Data <- read_csv('/content/Condensed Data.csv',show_col_types = FALSE)
Processed_Salaries <- read_excel("/content/Processed Salaries.xlsx")
NBA_Salaries <- read_excel("/content/NBA Salaries.xlsx")
PPG <- read_excel("/content/PPG.xlsx")
Tax_Threshold <- read_excel("/content/Tax Threshold.xlsx")
Team_Revenue <- read_excel("/content/Team Revenue.xlsx")
Team_Winning <- read_excel("/content/Team Winning.xlsx")
Cap_Minimum <- read_excel("/content/Cap Minimum.xlsx")
Team_Tax_Payments <- read_excel("/content/Team Tax Payments.xlsx")
```

# i. Condensed Data dataset
**Cap Minimum before detrending**

```r
# Removing missing records in the datasets if any
Condensed_Data = na.omit(Condensed_Data)
#names(Condensed_Data)
suppressWarnings({
  Condensed_Year.ts <- ts(Condensed_Data$`Cap Minimum`,frequency = 12, start=c(2013),end = c(2024))
  Condensed_Year.ts <- as.zoo(Condensed_Year.ts)
  main <- "Cap_Minimum Vs Time(Years)"
  xlab = 'Time'
  lty <- c("solid")

  plot(Condensed_Year.ts)
})
```

**Cap Minimum after detrending**

```r

# Detrending
diff.ts = diff(diff(Condensed_Year.ts))
plot(diff.ts,main = 'Cap Minimum over time (years)',
     xlab = 'Time',ylab = 'Detrended Cap Minimum')
```
